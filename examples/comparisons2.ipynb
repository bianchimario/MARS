```python
from downtime import load_dataset
import matplotlib.pyplot as plt
import lightgbm as lgb
from sklearn.metrics import classification_report
from MARS import MARS
import numpy as np
from datatime.conversion import awkward_to_flat
```

# ArticularyWordRecognition

```python
d = load_dataset("ArticularyWordRecognition")

X_train = d.X_train
X_test = d.X_test
y_train = d.y_train
y_test = d.y_test
```

```python
print(d)
```

## Multivariate Shapelets

```python
mars = MARS(num_shapelets=200, max_len=len(X_train[0][0]), min_len=10, seed=42, shapelet_indexes=False, indexes=False)

mars.fit(X_train)
```

```python
X_train_transformed = mars.transform(X_train)

X_test_transformed = mars.transform(X_test)

clf = lgb.LGBMClassifier()
clf.fit(X_train_transformed, y_train)
```

```python
y_pred_test = clf.predict(X_test_transformed)

print(classification_report(y_test, y_pred_test))
```

## Univariate Shapelets

```python
flat_X_train = awkward_to_flat(X_train).to_numpy()
flat_X_test = awkward_to_flat(X_test).to_numpy()
```

```python
# Putting each value inside a list

flat_X_train = [[x] for x in flat_X_train]
flat_X_test = [[x] for x in flat_X_test]
```

```python
mars = MARS(num_shapelets=200, max_len=len(X_train[0][0])*len(X_train[0]), min_len=10*len(X_train[0]), seed=42)

mars.fit(flat_X_train)
```

```python
X_train_transformed = mars.transform(flat_X_train)

X_test_transformed = mars.transform(flat_X_test)

clf = lgb.LGBMClassifier()
clf.fit(X_train_transformed, y_train)
```

```python
y_pred_test = clf.predict(X_test_transformed)

print(classification_report(y_test, y_pred_test))
```
